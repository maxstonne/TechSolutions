```python
!pip install numpy pandas matplotlib seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Carregar o conjunto de dados Iris
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

# Descrição estatística dos dados
desc_stats = data.describe()
print("Descrição Estatística:\n", desc_stats)

# Cálculo da moda
mode_stats = data.mode().iloc[0]
print("\nModa:\n", mode_stats)

# Distribuição dos dados
plt.figure(figsize=(12, 6))
for feature in iris.feature_names:
    sns.histplot(data[feature], kde=True, label=feature)
plt.legend()
plt.title('Distribuição das Características do Conjunto de Dados Iris')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()

# Boxplot para detectar outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=data.drop(columns='species'))
plt.title('Boxplot das Características do Conjunto de Dados Iris')
plt.xlabel('Características')
plt.ylabel('Valor')
plt.show()
```

```python
!pip install numpy pandas seaborn matplotlib

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o conjunto de dados Iris
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

# Calcular a matriz de correlação
correlation_matrix = data.corr()
print("Matriz de Correlação:\n", correlation_matrix)

# Visualização da matriz de correlação usando um mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de Calor da Matriz de Correlação')
plt.show()

# Scatterplot para visualizar a relação entre características
plt.figure(figsize=(12, 10))
sns.pairplot(data, hue='species', markers=["o", "s", "D"])
plt.suptitle('Pairplot das Características do Conjunto de Dados Iris', y=1.02)
plt.show()
```
